**Importing Required Libraries**

```python

import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.preprocessing import OneHotEncoder
```

**Loading Dataset**

```python
data=pd.read_csv("Churn_Modelling.csv")
data.head()
```

**Preprocessing Dataset**

```python
#Dropping Irrelevant Columns
data=data.drop(['RowNumber','CustomerId','Surname'],axis=1)
data.head()
```

**Encoding Categorical Values**

```python
#Encoding Gender Column 
gender_encoder=LabelEncoder()
data['Gender']=gender_encoder.fit_transform(data['Gender'])
data
```

```python
#Encoding Geography (OneHotEncoding)
geography_Onehotencoder=OneHotEncoder()
 =geography_Onehotencoder.fit_transform(data[['Geography']])

```

```python
geo_encoder
#Converting to Array 
geo_encoder.toarray()
```

```python
#Column Features 
geography_Onehotencoder.get_feature_names_out(['Geography'])
```

```python
#storing in DataFrame 
geo_encoded_df=pd.DataFrame(geo_encoder.toarray(),columns=geography_Onehotencoder.get_feature_names_out(['Geography']))
```

```python
geo_encoded_df
```

```python
#Combining With Original Dataset 
data=pd.concat([data.drop('Geography',axis=1),geo_encoded_df],axis=1) 
data.head()
```

**Saving Encoders**

```python
import pickle
```

```python
with open('geo_encoder.pkl','wb') as file:
    pickle.dump(geo_encoder,file)

with open('gender_encoder.pkl','wb') as file:
    pickle.dump(geo_encoder,file)
```

**Spliting Data**

```python
#Dividing in Dependent and Independent Features 
x=data.drop('Exited',axis=1)
y=data['Exited']
```

```python
#Train & Test Split
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#Scaling Data 
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```
