**Importing Required Libraries**

```python

import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.preprocessing import OneHotEncoder
```

**Loading Dataset**

```python
data=pd.read_csv("Churn_Modelling.csv")
data.head()
```

**Preprocessing Dataset**

```python
#Dropping Irrelevant Columns
data=data.drop(['RowNumber','CustomerId','Surname'],axis=1)
data.head()
```

**Encoding Categorical Values**

```python
#Encoding Gender Column 
gender_encoder=LabelEncoder()
data['Gender']=gender_encoder.fit_transform(data['Gender'])
data
```

```python
#Encoding Geography (OneHotEncoding)
geography_Onehotencoder=OneHotEncoder()
geo_encoder=geography_Onehotencoder.fit_transform(data[['Geography']])

```

```python
geo_encoder
#Converting to Array 
geo_encoder.toarray()
```

```python
#Column Features 
geography_Onehotencoder.get_feature_names_out(['Geography'])
```

```python
#storing in DataFrame 
geo_encoded_df=pd.DataFrame(geo_encoder.toarray(),columns=geography_Onehotencoder.get_feature_names_out(['Geography']))
```

```python
geo_encoded_df
```

```python
#Combining With Original Dataset 
data=pd.concat([data.drop('Geography',axis=1),geo_encoded_df],axis=1) 
data.head()
```

**Saving Encoders**

```python
import pickle
```

```python
with open('geo_encoder.pkl','wb') as file:
    pickle.dump(geo_encoder,file)

with open('gender_encoder.pkl','wb') as file:
    pickle.dump(geo_encoder,file)
```

**Spliting Data**

```python
#Dividing in Dependent and Independent Features 
x=data.drop('Exited',axis=1)
y=data['Exited']
```

```python
#Train & Test Split
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#Scaling Data 
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
#Saving Scaler

with open ('scaler.pkl','wb') as file:
    pickle.dump(scaler,file)
```

**AAN Implementation**

```python
# Importing Required Libraries 

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping, TensorBoard
import datetime
```

```python
#Building Our ANN 
model = Sequential([
    Dense(64, activation='relu', input_shape=(X_train.shape[1],)),  # Hidden Layer 01
    Dense(32, activation='relu'),  # Hidden Layer 02
    Dense(1, activation='sigmoid')  # Output Layer
])
```

```python
model.summary()
```

```python
#setting Parameters for model 
import tensorflow
opt=tensorflow.keras.optimizers.Adam(learning_rate=0.01)
loss=tensorflow.keras.losses.BinaryCrossentropy()
```

```python
#Compiling the model 
model.compile(optimizer=opt,loss=loss,metrics=['accuracy'])
```

```python
#Setup logging for model training 
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
log_dir='logs/fit/'+ datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorflow_callbacks=TensorBoard(log_dir=log_dir,histogram_freq=1)
```

```python
#Setup Early Stopping 
early_stopping_callbacks=EarlyStopping(monitor='val_loss',patience=10,restore_best_weights=True)
```

**Training Model**

```python
#Training Modle 
history=model.fit(
    X_train,Y_train,validation_data=(X_test,Y_test),epochs=100,
    callbacks=[tensorflow_callbacks,early_stopping_callbacks]
)
```

```python
#saving the model 
model.save('modle.h5')
```
